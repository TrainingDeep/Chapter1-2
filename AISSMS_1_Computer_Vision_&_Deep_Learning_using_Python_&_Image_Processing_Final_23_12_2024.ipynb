```python
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
img = imread('/content/brics.jpeg')
img_new = rgb2gray(img)

plt.subplot(1,2,1), imshow(img)
plt.title('RGB Format')

plt.subplot(122), imshow(img_new)
plt.title('Grayscale Format')

plt.show()
```

```python
from skimage.transform import resize
img = imread('/content/brics.jpeg')
#resize image
img_resized = resize(img, (300, 300))

#plot images
plt.subplot(121), imshow(img)
plt.title('Original Image')
plt.subplot(122), imshow(img_resized)
plt.title('Resized Image')
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image
img = cv2.imread('/content/Amitabh.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian Blur to reduce noise
img_blur = cv2.GaussianBlur(img, (3,3), 0)

# Canny Edge Detection
edges = cv2.Canny(img_blur, 100, 200)

# Display the original image and the edges
cv2_imshow(edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
